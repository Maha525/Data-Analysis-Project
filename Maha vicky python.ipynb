##  Data cleaning – Video Uploads & Subscribers

```python
import pandas as pd
df = pd.read_csv("top-5000-youtube-channels.csv", encoding='latin1')


df['Video Uploads'] = pd.to_numeric(df['Video Uploads'], errors='coerce')
df['Subscribers'] = pd.to_numeric(df['Subscribers'], errors='coerce')
df['Video views'] = pd.to_numeric(df['Video views'], errors='coerce')
```

## 
Data cleaning – Grade column

```python
df['Grade'] = df['Grade'].astype(str)
df
```

##  Find average views for each channel

```python
df['Average_Views'] = df['Video views'] / df['Video Uploads']
df[['Channel name','Average_Views']].head()
```

##  Top 5 channels with maximum uploads

```python
df.nlargest(5, 'Video Uploads')[['Channel name','Video Uploads']]
```

## Subscribers Distribution (Histogram)

```python

import matplotlib.pyplot as plt
df = pd.read_csv("top-5000-youtube-channels.csv", encoding='latin1')

plt.figure(figsize=(10,5))
plt.hist(df['Video views'].dropna(), bins=50)
plt.xlabel("Views")
plt.ylabel("Count")
plt.title("Distribution of Video Views")
plt.show

```

##  Top 10 Channels by Subscribers (Bar Chart)

```python
top10 = df.nlargest(10, 'Subscribers')

plt.figure(figsize=(12,6))
plt.bar(top10['Channel name'], top10['Subscribers'])
plt.xticks(rotation=90)
plt.xlabel("Channel Name")
plt.ylabel("Subscribers")
plt.title("Top 10 Channels by Subscribers")
plt.show()
```

##  Top 10 Channels by Video Views (Bar Chart)

```python
plt.figure(figsize=(12,6))
top10_views = df.nlargest(10, 'Video views')
plt.bar(top10_views['Channel name'], top10_views['Video views'], color=plt.cm.Set2.colors)
plt.xticks(rotation=90)
plt.title("Top 10 Channels by Video Views")
plt.ylabel("Video Views")
plt.show()
```

 ##  Scatter Plot (Subscribers vs Views)

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Subscribers'], df['Video views'], color='blue')
plt.xlabel("Subscribers")
plt.ylabel("Video Views")
plt.title("Subscribers vs Views")
plt.show()
```

##  Grade-wise Average Views (Bar Chart)

```python
plt.figure(figsize=(8,5))
avg_views = df.groupby('Grade')['Video views'].mean()

plt.bar(avg_views.index, avg_views.values, color=plt.cm.Set3.colors)
plt.xlabel("Grade")
plt.ylabel("Average Views")
plt.title("Grade-wise Average Views")
plt.show()
```

##  Grade-wise Subscribers (Bar Chart)

```python
plt.figure(figsize=(8,5))
avg_subs = df.groupby('Grade')['Subscribers'].mean()

plt.bar(avg_subs.index, avg_subs.values, color=plt.cm.Pastel1.colors)
plt.xlabel("Grade")
plt.ylabel("Average Subscribers")
plt.title("Grade-wise Subscribers")
plt.show()
```
